# Import


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.linear_model import LinearRegression
```

#  Load dataset

```python
# ## Load dataset
boston_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
boston_df = pd.read_csv(boston_url)
```

# Display basic info and first rows

```python
# ## Display basic info and first rows
print(boston_df.info())
print(boston_df.head())
```

# Visualization: Boxplot of MEDV

```python
# ## Visualization: Boxplot of MEDV
plt.figure(figsize=(8,5))
sns.boxplot(y=boston_df['MEDV'])
plt.title('Boxplot of Median Value of Owner-Occupied Homes (MEDV)')
plt.show()
```

# Visualization: Histogram of PTRATIO

```python
# ## Visualization: Histogram of PTRATIO
plt.figure(figsize=(8,5))
sns.histplot(boston_df['PTRATIO'], bins=20, kde=True)
plt.title('Histogram of Pupil-Teacher Ratio (PTRATIO)')
plt.show()
```

# Visualization: Scatter plot NOX vs INDUS

```python
# ## Visualization: Scatter plot NOX vs INDUS
plt.figure(figsize=(8,5))
sns.scatterplot(x=boston_df['NOX'], y=boston_df['INDUS'])
plt.title('Scatter Plot: NOX vs INDUS')
plt.xlabel('Nitric Oxide Concentration (NOX)')
plt.ylabel('Non-Retail Business Acreage (INDUS)')
plt.show()
```

# Statistical Test: Independent T-test for CHAS

```python
# ## Statistical Test: Independent T-test for CHAS
chas1 = boston_df[boston_df['CHAS'] == 1]['MEDV']
chas0 = boston_df[boston_df['CHAS'] == 0]['MEDV']
t_stat, p_val = stats.ttest_ind(chas1, chas0)
print(f'T-test: t-statistic = {t_stat}, p-value = {p_val}')
```

# Statistical Test: Pearson correlation between NOX and INDUS

```python
# ## Statistical Test: Pearson correlation between NOX and INDUS
corr_coeff, p_value = stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
print(f'Pearson Correlation: Coefficient = {corr_coeff}, p-value = {p_value}')
```

# Regression Analysis: DIS impact on MEDV

```python
# ## Regression Analysis: DIS impact on MEDV
X = boston_df[['DIS']]
y = boston_df['MEDV']
model = LinearRegression()
model.fit(X, y)
```

# Print regression results

```python
# Print regression results
print(f'Intercept: {model.intercept_}')
print(f'Coefficient: {model.coef_[0]}')
print(f'R-squared: {model.score(X, y)}')
```
